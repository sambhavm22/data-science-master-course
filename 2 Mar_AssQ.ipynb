### Answer 2

A scatter plot is a type of plot in data visualization that uses a collection of points to represent the values of two variables for a set of data. Each point in the plot represents a single observation in the dataset and is plotted using the values of the two variables on the x and y-axes.

```python
np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(6,3))
plt.scatter(x,y)
plt.xlabel('represnting x values')
plt.ylabel('represnting y values')
plt.title('plot btw x vs y')
plt.show()
```

### Answer 3

The subplot() function in matplotlib.pyplot is used to create multiple plots (subplots) within a single figure. It allows users to create a grid of subplots of a given size and arrangement, and to plot different data or types of plots in each subplot.

The subplot() function takes three integer arguments: the number of rows of subplots, the number of columns of subplots, and the index of the subplot to create. 


```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(8, 8))

# Plot the four lines in the subplots
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Add titles and axis labels to the subplots
axs[0, 0].set_title('Line 1')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')

axs[0, 1].set_title('Line 2')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')

axs[1, 0].set_title('Line 3')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')

axs[1, 1].set_title('Line 4')
axs[1, 1].set_xlabel('X')
axs[1, 1].set_ylabel('Y')

# Show the plot
plt.show()

```

### Answer 4

A bar plot (also called a bar chart) is a graphical representation of data in which bars of rectangular shapes are used to represent the values of different categories. The length or height of each bar corresponds to the value of the data it represents. Bar plots are commonly used to compare the values of different categories or groups of data.

Bar plots are used to visualize and compare categorical data, such as counts, frequencies, or proportions of different categories or groups. They are particularly useful when the data being compared are discrete, as the bars can represent distinct categories or groups. Bar plots are often used to show the distribution of data across different categories, or to highlight differences between groups of data.

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(6,3))
plt.bar(company,profit)
plt.xlabel('company')
plt.ylabel('profit')
plt.title('profit of companies')
```

### Answer 5

A box plot (also called a box-and-whisker plot) is a graphical representation of data that shows the distribution of a dataset along a number line. It displays the range, median, quartiles, and any outliers of a dataset. The box plot is commonly used to represent the distribution of continuous data and identify any outliers in the data.

```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```

```python
box1
```

```python
box2
```

```python
data = [box1, box2]

plt.boxplot(data, labels=["Box 1", "Box 2"])
plt.title("Box Plot")
plt.ylabel("Values")

plt.show()
```
