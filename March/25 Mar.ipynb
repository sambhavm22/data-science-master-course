### Impact of outlier in our analysis

Skewness: Outliers can cause the distribution of flight prices to be skewed towards one direction. This can affect our ability to analyze the central tendency of the data accurately. For example, if a few flights have significantly higher prices than the rest, the mean price of flights can be much higher than the median price, which may not be representative of the overall trend.

Variability: Outliers can increase the variability or spread of the data, which can affect our ability to estimate the standard deviation and other measures of variability accurately. For example, if a few flights have very high prices, this can cause the standard deviation to be overestimated, which can impact our confidence in the statistical significance of our analysis.

Model assumptions: Outliers can violate the assumptions of some statistical models. For example, linear regression models assume that the data points are normally distributed with equal variance. If there are outliers, this assumption may be violated, which can lead to inaccurate predictions and incorrect conclusions.

Sampling bias: Outliers can also result from sampling bias, which can occur when the data is not collected randomly or is not representative of the population of interest. In this case, outliers can significantly impact our ability to generalize the findings to the population.

### Q6. You are working for a travel agency, and your boss has asked you to analyze the Flight Price dataset to identify the peak travel season. What features would you analyze to identify the peak season, and how would you present your findings to your boss?


```python
flight['Date_of_Journey'] = pd.to_datetime(flight['Date_of_Journey'], dayfirst=True)
```

```python
flight['Month'] = flight['Date_of_Journey'].dt.month
flight['Quarter'] = flight['Date_of_Journey'].dt.quarter
flight['Season'] = (flight['Month'] % 12 + 3) // 3
```

```python
price_by_month = flight.groupby('Month')['Price'].mean()
price_by_quarter = flight.groupby('Quarter')['Price'].mean()
price_by_season = flight.groupby('Season')['Price'].mean()
```

```python
price_by_destination = flight.groupby('Destination')['Price'].mean()
price_by_duration = flight.groupby('Duration')['Price'].mean()
price_by_airline = flight.groupby('Airline')['Price'].mean()
price_by_route = flight.groupby('Route')['Price'].mean()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

ax[0, 0].plot(price_by_month.index, price_by_month.values)
ax[0, 0].set_xlabel('Month')
ax[0, 0].set_ylabel('Average Price')
ax[0, 0].set_title('Flight Prices by Month')

ax[0, 1].plot(price_by_quarter.index, price_by_quarter.values)
ax[0, 1].set_xlabel('Quarter')
ax[0, 1].set_ylabel('Average Price')
ax[0, 1].set_title('Flight Prices by Quarter')

ax[1, 0].bar(price_by_season.index, price_by_season.values)
ax[1, 0].set_xlabel('Season')
ax[1, 0].set_ylabel('Average Price')
ax[1, 0].set_title('Flight Prices by Season')

ax[1, 1].bar(price_by_destination.index, price_by_destination.values)
ax[1, 1].set_xlabel('Destination')
ax[1, 1].set_ylabel('Average Price')
ax[1, 1].set_title('Flight Prices by Destination')

plt.tight_layout()
plt.show()
```

```python
flight.groupby(flight['Date_of_Journey'].dt.month_name())['Price'].sum().plot.bar(title='Revenue per Months')
```

### Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python
df['Size'] = df['Size'].str.replace('M','000')
df['Size'] = df['Size'].str.replace('k','')
df['Size'] = df['Size'].str.replace('+','')
df['Size'] = df['Size'].replace('Varies with device', np.nan)
df['Size'] = df['Size'].dropna()
df['Size'] = df['Size'].astype(float)
```
