```python
### Answer 3

import math

def normal_pdf(x, mean, std_dev):
    """
    Calculates the probability density function of a normal distribution
    at a given point x, with mean mean and standard deviation std_dev.
    """
    exponent = -(x - mean)**2 / (2 * std_dev**2)
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    return coefficient * math.exp(exponent)
```

```python
pdf = normal_pdf(1, 0, 1)
print(pdf)
```

2) Quality control: In a manufacturing process, a sample of products is inspected and the number of defective products (failure) is recorded. The binomial distribution can be used to model the probability of a certain number of defective products in a fixed sample size.

```python
### Answer 6

import math

def poisson_cdf(x, mean):
    """
    Calculates the cumulative distribution function of a Poisson distribution
    at a given point x, with mean mean.
    """
    cdf = 0
    for i in range(x + 1):
        cdf += math.exp(-mean) * mean**i / math.factorial(i)
    return cdf
```

```python
cdf = poisson_cdf(3, 2)
print(cdf)
```

### Answer 7

Binomial and Poisson distributions are two different types of probability distributions that are used to model the number of events occurring in a fixed number of trials or a fixed unit of time. While both distributions are used to model count data, they have some fundamental differences:

Number of Trials: Binomial distribution is used when the number of trials is fixed and finite, while Poisson distribution is used when the number of events can theoretically be infinite.

Probability of Success: In the binomial distribution, the probability of success (i.e., the probability that the event of interest occurs on any given trial) remains constant across all trials, whereas in the Poisson distribution, the probability of occurrence of an event is assumed to be constant, but the probability of success is very low when the event rate is low.

Relationship between Mean and Variance: In the binomial distribution, the mean and variance are both equal to n*p (where n is the number of trials and p is the probability of success), whereas in the Poisson distribution, the mean and variance are both equal to lambda (the rate parameter).

Application: The binomial distribution is commonly used to model the number of successes in a fixed number of independent trials, whereas the Poisson distribution is used to model the number of rare events that occur in a fixed unit of time or space.

### Answer 10

In a normal distribution, the least frequent data (i.e., the data with the lowest probability density) appears in the tails of the distribution, which are the regions furthest away from the mean. Specifically, the data in the tails that are more than 2 or 3 standard deviations away from the mean are the least frequent.
