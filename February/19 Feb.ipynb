### Answer 3

App routing is the process of mapping URL routes to specific functions in your Flask application. This is achieved using the @app.route decorator, which is used to define a route in your application. 

We use app routes in Flask for several reasons:

To provide a clean and organized way to define the URL endpoints of our application.

To map a URL endpoint to a specific function in our Flask application.

To make it easy to add new URL endpoints to our application as we build out new features.

To allow for dynamic URL routing based on user input or other factors.

To provide a way to handle HTTP methods like GET, POST, PUT, DELETE, etc.

```python
# Answer 4

!pip install flask
from flask import Flask
app = Flask(__name__)

@app.route('/welcome')
def welcome_message():
    return "Welcome to ABC Corporation"

@app.route('/')
def details():
    return "Company Name: ABC Corporation, Location: India, Contact Detail: 999-999-9999"

if __name__=="__main__":
    app.run(host="0.0.0.0")  

    
    
```

### Answer 5

In Flask, the url_for() function is used for URL building. The url_for() function generates a URL to a specific function or endpoint, based on the endpoint name provided as an argument.



```python

from flask import Flask, url_for
from flask import request

app = Flask(__name__)

@app.route('/')
def home():
    return "welcome to flask app!"

@app.route('/profile/<username>')
def profile(username):
    return 'Hello, {}!'.format(username)

with app.test_request_context():
    print(url_for('home'))  # Output: '/'
    print(url_for('profile', username='sam')) 

if __name__=="__main__":
    app.run(host="0.0.0.0")    
```
