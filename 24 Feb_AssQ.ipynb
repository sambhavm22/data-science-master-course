```python

import pandas as pd

def add_weekday_column(df, date_column_name='Date', weekday_column_name='Weekday'):
    
    df[weekday_column_name] = df[date_column_name].dt.day_name()
    
    return df
```

```python
# create a sample Pandas DataFrame
df = pd.DataFrame({
    "Date": pd.date_range(start='2023-01-01', periods=7, freq='D')
})

# call the function to add the weekday column
df = add_weekday_column(df)

# print the modified DataFrame
print(df)
```
